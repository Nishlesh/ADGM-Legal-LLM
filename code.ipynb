```python
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from google.auth.transport.requests import Request
from googleapiclient.discovery import build
import os

# If modifying these scopes, delete the token file.
SCOPES = ['https://www.googleapis.com/auth/drive.metadata.readonly'] # Recommended scope for listing links

# Define paths for your credential files
CLIENT_SECRETS_FILE = '/home/user/legal_llm/client_secret.json' # Your "credentials.json" renamed
TOKEN_FILE = '/home/user/legal_llm/application_default_credentials.json' # Your "application_default_credentials.json"

def get_drive_service():
    creds = None
    # Attempt to load credentials from the stored token file
    if os.path.exists(TOKEN_FILE):
        creds = Credentials.from_authorized_user_file(TOKEN_FILE, SCOPES)

    # If no valid credentials, or if they are expired and can't be refreshed,
    # initiate a new OAuth flow using the client secrets file.
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            # Attempt to refresh the token
            try:
                creds.refresh(Request())
            except Exception as e:
                print(f"Error refreshing token: {e}. Initiating new flow.")
                creds = None # Force a new flow if refresh fails
        
        if not creds: # If no creds or refresh failed, start a new flow
            print("No valid credentials found or token refresh failed. Initiating new authentication flow...")
            flow = InstalledAppFlow.from_client_secrets_file(
                CLIENT_SECRETS_FILE, SCOPES)
            creds = flow.run_local_server(port=0)
            # Save the new credentials (including refresh token) for the next run
            with open(TOKEN_FILE, 'w') as token:
                token.write(creds.to_json())
    
    service = build('drive', 'v3', credentials=creds)
    return service

def list_files_in_folder(service, folder_id):
    """
    Lists all files in a specific Google Drive folder and returns their webViewLink.
    """
    file_links = []
    page_token = None
    while True:
        try:
            results = service.files().list(
                q=f"'{folder_id}' in parents and trashed = false",
                fields="nextPageToken, files(id, name, webViewLink, mimeType, capabilities/canDownload)", # Added mimeType and canDownload for better info
                pageToken=page_token
            ).execute()
            items = results.get('files', [])

            if not items:
                print('No files found in the folder.')
                break
            for item in items:
                # Google Docs, Sheets, Slides don't have a direct "download" link like regular files.
                # webViewLink is generally what you want for RAG if you intend to read content via a web view.
                # If you need to *download* specific file types, you'd use service.files().export()
                # for Google Workspace files, or service.files().get(fileId=item['id'], alt='media') for others.
                print(f"Found file: {item['name']} (ID: {item['id']}) - Type: {item['mimeType']} - Link: {item['webViewLink']}")
                
                # Filter for files that are likely documents and can be viewed or downloaded
                # You might want to refine this based on what 'rag.import_files' supports
                if item['mimeType'] != 'application/vnd.google-apps.folder': # Exclude folders
                     file_links.append(item['webViewLink'])
            
            page_token = results.get('nextPageToken', None)
            if not page_token:
                break
        except Exception as e:
            print(f"An error occurred while listing files: {e}")
            break # Exit loop on error
    return file_links

if __name__ == '__main__':
    # Replace with the actual ID of your Google Drive folder
    # You can find the folder ID in the URL when you open the folder in Google Drive.
    DRIVE_FOLDER_ID = "1Js2hPMLPpagi1HKHBHlaO2gvwW2hdgBa"

    try:
        drive_service = get_drive_service()
        all_file_links = list_files_in_folder(drive_service, DRIVE_FOLDER_ID)

        print("\n--- All File Links Retrieved ---")
        if all_file_links:
            for link in all_file_links:
                print(link)
        else:
            print("No suitable file links found.")

        # Example of how you would integrate it back into your RAG code:
        # import rag # Assuming rag is available from your environment
        # rag_corpus_name = "your_rag_corpus_name_here" # Replace with your actual corpus name
        # rag.import_files(
        #     rag_corpus_name, # Use the actual corpus name here
        #     all_file_links,
        #     transformation_config=rag.TransformationConfig(
        #         chunking_config=rag.ChunkingConfig(
        #             chunk_size=512,
        #             chunk_overlap=100,
        #         ),
        #     ),
        #     max_embedding_requests_per_min=1000  # Optional
        # )

    except Exception as e:
        print(f"An unhandled error occurred in the main execution block: {e}")
```
